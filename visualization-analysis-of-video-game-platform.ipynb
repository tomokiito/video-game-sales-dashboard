### Task 1 Prototype - Analysis of Platform Market Share Trends

```python
import pandas as pd
import matplotlib.pyplot as plt

# Loading the dataset
file_path = '/kaggle/input/video-game-sales-with-ratings/Video_Games_Sales_as_at_22_Dec_2016.csv'
data = pd.read_csv(file_path)

# Filtering data to include only up to 2016
data = data.dropna(subset=['Year_of_Release'])
data = data[data['Year_of_Release'] <= 2016]

# Calculating total sales per year
total_sales_per_year = data.groupby('Year_of_Release')['Global_Sales'].sum()

# Calculating platform sales per year, normalized by total yearly sales
platform_sales_by_year = data.groupby(['Year_of_Release', 'Platform'])['Global_Sales'].sum()
platform_sales_share = platform_sales_by_year.div(total_sales_per_year, level='Year_of_Release')

# Unstacking for easier plotting
platform_sales_share_unstacked = platform_sales_share.unstack().fillna(0)

# Selecting top 10 platforms based on total global sales
top_10_platforms = platform_sales_by_year.groupby('Platform').sum().nlargest(10).index
top_10_platform_sales_share = platform_sales_share_unstacked[top_10_platforms]

# Plotting top 10 platforms trends (multiplied by 100 for percentage)
(top_10_platform_sales_share * 100).plot(kind='area', stacked=True, figsize=(14, 8), colormap='viridis')
plt.title("Top 10 Platform Popularity Trends by Year (Normalized by Yearly Total Sales)", fontsize=16)
plt.xlabel("Year of Release", fontsize=14)
plt.ylabel("Platform Sales Share (%)", fontsize=14)
plt.legend(title="Platform", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.show()

# Grouping platforms by company
platform_groups = {
    'Nintendo': ['Wii', 'WiiU', 'DS', '3DS', 'GC', 'NES', 'SNES', 'N64', 'GB', 'GBA'],
    'Sony': ['PS', 'PS2', 'PS3', 'PS4', 'PSP', 'PSV'],
    'Microsoft': ['XB', 'X360', 'XOne']
}

# Aggregating sales shares by company
grouped_sales_share = pd.DataFrame(index=platform_sales_share_unstacked.index)
for group, platforms in platform_groups.items():
    grouped_sales_share[group] = platform_sales_share_unstacked[platforms].sum(axis=1)

# Plotting grouped company trends (multiplied by 100 for percentage)
(grouped_sales_share * 100).plot(kind='area', stacked=True, figsize=(14, 8), colormap='viridis')
plt.title("Platform Popularity Trends by Year (Grouped by Company, Normalized by Yearly Total Sales)", fontsize=16)
plt.xlabel("Year of Release", fontsize=14)
plt.ylabel("Sales Share (%)", fontsize=14)
plt.legend(title="Company", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.show()
```

### Task 2 Prototype - Identifying Key Factors Influencing Global Sales

```python
# Importing necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder, StandardScaler

# Handling missing values
data['User_Score'] = pd.to_numeric(data['User_Score'], errors='coerce')
data['Critic_Score'] = data['Critic_Score'].fillna(data['Critic_Score'].mean())
data['User_Score'] = data['User_Score'].fillna(data['User_Score'].mean())
data = data.dropna(subset=['Year_of_Release'])

# Encoding categorical features (if needed for later modeling)
categorical_features = ['Platform', 'Genre', 'Publisher', 'Developer', 'Rating']
label_encoders = {}
for feature in categorical_features:
    label_encoders[feature] = LabelEncoder()
    data[f'{feature}_Encoded'] = label_encoders[feature].fit_transform(data[feature].astype(str))

# Scaling numerical features (excluding Global_Sales)
scaler = StandardScaler()
data[['Critic_Score', 'User_Score']] = scaler.fit_transform(data[['Critic_Score', 'User_Score']])

# --- Visualization without scaling Global_Sales ---

# Impact of Platform on Global Sales
plt.figure(figsize=(14, 6))
sns.boxplot(data=data, x='Platform', y='Global_Sales', palette="Set3")
plt.title("Impact of Platform on Global Sales")
plt.xlabel("Platform")
plt.ylabel("Global Sales (Millions)")
plt.xticks(rotation=45)
plt.grid(True)
plt.show()

# Impact of Genre on Global Sales
plt.figure(figsize=(14, 6))
sns.boxplot(data=data, x='Genre', y='Global_Sales', palette="Set2")
plt.title("Impact of Genre on Global Sales")
plt.xlabel("Genre")
plt.ylabel("Global Sales (Millions)")
plt.xticks(rotation=45)
plt.grid(True)
plt.show()

# Impact of Rating on Global Sales
plt.figure(figsize=(10, 6))
sns.boxplot(data=data, x='Rating', y='Global_Sales', palette="Set1")
plt.title("Impact of Rating on Global Sales")
plt.xlabel("Rating")
plt.ylabel("Global Sales (Millions)")
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

### Task 3 Prototype - Predicting Future Platform Popularity

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Calculate total sales and sales share by platform for each year
total_sales_per_year = data.groupby('Year_of_Release')['Global_Sales'].sum()
platform_sales = data.groupby(['Year_of_Release', 'Platform'])['Global_Sales'].sum()
platform_sales_share = platform_sales.div(total_sales_per_year, level='Year_of_Release').unstack().fillna(0)

platform_groups = {
    'Nintendo': ['Wii', 'WiiU', 'DS', '3DS', 'GC', 'NES', 'SNES', 'N64', 'GB', 'GBA'],
    'Sony': ['PS', 'PS2', 'PS3', 'PS4', 'PSP', 'PSV'],
    'Microsoft': ['XB', 'X360', 'XOne']
}

company_sales_share = pd.DataFrame(index=platform_sales_share.index)
for company, platforms in platform_groups.items():
    company_sales_share[company] = platform_sales_share[platforms].sum(axis=1)

# Linear regression (just simple predictions up to 2020)
future_years = np.arange(2016, 2021).reshape(-1, 1)
historical_years = company_sales_share.index.values.reshape(-1, 1)

plt.figure(figsize=(14, 8))

# Create a model and plot for each company
for company in company_sales_share.columns:
    sales_data = company_sales_share[company].values * 100
    model = LinearRegression().fit(historical_years, sales_data)
    future_pred = model.predict(future_years)

    # Historical data plot
    plt.plot(company_sales_share.index, sales_data, marker='o', linewidth=2, label=f'{company} (historical)')
    # Predicted data plot
    plt.plot(future_years.flatten(), future_pred, linestyle='--', linewidth=2, label=f'{company} (predicted)')

plt.title('Simple Linear Prediction of Platform Popularity (Company-wise)', fontsize=16)
plt.xlabel('Year', fontsize=14)
plt.ylabel('Platform Sales Share (%)', fontsize=14)  
plt.legend()
plt.grid(True)
plt.show()
```

## Part 4ï¼šData Visualization 

```python
import altair as alt

# Calculate total global sales per year
total_sales_per_year = data.groupby('Year_of_Release')['Global_Sales'].sum().reset_index()

# Calculate platform sales per year
platform_sales = data.groupby(['Year_of_Release', 'Platform'])['Global_Sales'].sum().reset_index()

# Calculate market share (%) for each platform per year
platform_sales = platform_sales.merge(total_sales_per_year, on='Year_of_Release', suffixes=('', '_Total'))
platform_sales['Sales_Share'] = (platform_sales['Global_Sales'] / platform_sales['Global_Sales_Total']) * 100

# Define platform categories (Nintendo, Sony, Microsoft)
def platform_company(platform):
    nintendo = ['Wii', 'WiiU', 'DS', '3DS', 'GC', 'NES', 'SNES', 'N64', 'GB', 'GBA']
    sony = ['PS', 'PS2', 'PS3', 'PS4', 'PSP', 'PSV']
    microsoft = ['XB', 'X360', 'XOne']
    if platform in nintendo:
        return 'Nintendo'
    elif platform in sony:
        return 'Sony'
    elif platform in microsoft:
        return 'Microsoft'
    else:
        return 'Other'

platform_sales['Company'] = platform_sales['Platform'].apply(platform_company)

# Aggregate market share by company per year
company_sales_share = platform_sales.groupby(['Year_of_Release', 'Company'])['Sales_Share'].sum().reset_index()

# Create an interactive stacked area chart using Altair
chart = alt.Chart(company_sales_share).mark_area(opacity=0.7).encode(
    x=alt.X('Year_of_Release:O', title='Year of Release'),
    y=alt.Y('Sales_Share:Q', stack='normalize', title='Market Share (%)'),
    color=alt.Color('Company:N', scale=alt.Scale(scheme='category20'), title='Company'),
    tooltip=[
        alt.Tooltip('Year_of_Release:O', title='Year'),
        alt.Tooltip('Company:N', title='Company'),
        alt.Tooltip('Sales_Share:Q', title='Sales Share (%)', format='.2f')
    ]
).properties(
    width=700,
    height=400,
    title='Interactive Visualization: Platform Market Share Trends by Company'
).interactive()

# Display the chart
# chart
```

```python
import vegafusion as vf

alt.data_transformers.enable("vegafusion")

# Prepare data for visualization
platform_data = data[['Platform', 'Global_Sales']].rename(columns={'Platform': 'Category_Value'})
platform_data['Category'] = 'Platform'

genre_data = data[['Genre', 'Global_Sales']].rename(columns={'Genre': 'Category_Value'})
genre_data['Category'] = 'Genre'

rating_data = data[['Rating', 'Global_Sales']].rename(columns={'Rating': 'Category_Value'})
rating_data['Category'] = 'Rating'

# Combine data
combined_data = pd.concat([platform_data, genre_data, rating_data])

# Filter top 10 categories for each category type
top_platforms = data.groupby('Platform')['Global_Sales'].sum().nlargest(10).index
top_genres = data.groupby('Genre')['Global_Sales'].sum().nlargest(10).index
top_ratings = data.groupby('Rating')['Global_Sales'].sum().nlargest(10).index

combined_data = combined_data[
    ((combined_data['Category']=='Platform') & (combined_data['Category_Value'].isin(top_platforms))) |
    ((combined_data['Category']=='Genre') & (combined_data['Category_Value'].isin(top_genres))) |
    ((combined_data['Category']=='Rating') & (combined_data['Category_Value'].isin(top_ratings)))
]

# Calculate category order by total sales
category_order = combined_data.groupby('Category_Value')['Global_Sales'].sum().sort_values(ascending=False).index.tolist()

# Selection parameter
category_param = alt.param(
    bind=alt.binding_select(options=['Platform', 'Genre', 'Rating'], name='Category:'),
    value='Platform'
)

# Violin plot with custom sorting and zoomed initial scale
violin = alt.Chart(combined_data).transform_filter(
    alt.datum.Category == category_param
).transform_density(
    'Global_Sales',
    as_=['Global_Sales', 'Density'],
    groupby=['Category_Value'],
    extent=[0, combined_data['Global_Sales'].quantile(0.95)],  # zoom into 95 percentile
    counts=True,
    steps=200
).mark_area(orient='horizontal', opacity=0.7).encode(
    y=alt.Y('Global_Sales:Q', title='Global Sales (Millions)', 
            scale=alt.Scale(domain=[0, combined_data['Global_Sales'].quantile(0.95)])),  # zoomed scale
    x=alt.X('Density:Q',
            stack='center',
            impute=None,
            title=None,
            axis=alt.Axis(labels=False, values=[0], grid=False, ticks=True)),
    color=alt.Color('Category_Value:N', legend=None),
    column=alt.Column('Category_Value:N',
                      header=alt.Header(title='', labelOrient='bottom'),
                      sort=category_order),  # custom sorting from left (highest) to right (lowest)
    tooltip=[
        alt.Tooltip('Category_Value:N', title='Category'),
        alt.Tooltip('Global_Sales:Q', title='Global Sales')
    ]
).properties(
    width=100,
    height=400,
    title='Interactive Violin Plot of Global Sales by Category (Sorted & Zoomed)'
).add_params(
    category_param
).interactive()

# violin
```

```python
# Calculate yearly total sales
total_sales_per_year = data.groupby('Year_of_Release')['Global_Sales'].sum()

# Calculate yearly sales by platform, normalize to get share
platform_sales = data.groupby(['Year_of_Release', 'Platform'])['Global_Sales'].sum()
platform_sales_share = platform_sales.div(total_sales_per_year, level='Year_of_Release').unstack().fillna(0)

# Group by company
company_sales_share = pd.DataFrame(index=platform_sales_share.index)
for company, platforms in platform_groups.items():
    company_sales_share[company] = platform_sales_share[platforms].sum(axis=1)

# Prepare forecast data (Linear Regression)
forecast_years = np.arange(2016, 2021).reshape(-1, 1)
historical_years = company_sales_share.index.values.reshape(-1, 1)

forecast_df = pd.DataFrame()

for company in company_sales_share.columns:
    model = LinearRegression().fit(historical_years, company_sales_share[company].values)
    predicted = model.predict(forecast_years)

    # Create dataframe for forecast
    temp_df = pd.DataFrame({
        'Year': forecast_years.flatten(),
        'Company': company,
        'Sales_Share': predicted,
        'Type': 'Predicted'
    })
    forecast_df = pd.concat([forecast_df, temp_df])

# Historical data
historical_df = company_sales_share.reset_index().melt(id_vars='Year_of_Release', var_name='Company', value_name='Sales_Share')
historical_df['Type'] = 'Actual'
historical_df = historical_df.rename(columns={'Year_of_Release': 'Year'})

# Combine actual + forecast
full_df = pd.concat([historical_df, forecast_df], ignore_index=True)
full_df['Sales_Share'] = full_df['Sales_Share'] * 100  # Convert to percentage

# Altair line chart
chart = alt.Chart(full_df).mark_line(point=True).encode(
    x=alt.X('Year:O', title='Year'),
    y=alt.Y('Sales_Share:Q', title='Platform Sales Share (%)'),
    color=alt.Color('Company:N', title='Company'),
    strokeDash=alt.StrokeDash('Type:N', title='Data Type'),  # Solid for actual, dashed for predicted
    tooltip=[
        alt.Tooltip('Year:O'),
        alt.Tooltip('Company:N'),
        alt.Tooltip('Sales_Share:Q', format='.2f'),
        alt.Tooltip('Type:N')
    ]
).properties(
    width=700,
    height=400,
    title='Platform Popularity Forecast (Actual vs Predicted)'
).interactive()

# chart
```
